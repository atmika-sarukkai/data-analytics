```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
months = ['Feb', 'March', 'Apr']
person_MAD_twitter = [0.753969, 0.614207, 0.578458]
person_MAD_sfpd = [0.881810, 0.890388, 0.827001]

property_MAD_twitter = [1.267038, 0.457822, 0.532316]
property_MAD_sfpd = [0.843880, 0.859586, 0.791529]

drug_MAD_twitter = [0.869728, 0.906365, 0.289929]
drug_MAD_sfpd = [0.590805, 0.476577, 0.465612]

minor_MAD_twitter = [0.529065, 0.620861, 0.295713]
minor_MAD_sfpd = [0.762669, 0.767644, 0.694062]
```

```python
person_slope_twitter = [0.00432172,-0.00127345, 0.00099727]
person_slope_sfpd = [0.0033085, -0.01103973, -0.00822429]

property_slope_twitter = [-0.00850907,-0.00420792, -0.00312265]
property_slope_sfpd = [-0.00172882, -0.00925167, -0.03380073]

drug_slope_twitter = [-0.00092631,-0.00615187, -0.00140617]
drug_slope_sfpd = [-0.03441307, 0.00368361, 0.00714298]

minor_slope_twitter = [-0.00100104, 0., -0.00322676]
minor_slope_sfpd = [-0.00325252, -0.00482895, -0.01182]
```

```python
person_crime_stats = pd.DataFrame({'Month' : months})
person_crime_stats['Twitter MAD'] = person_MAD_twitter
person_crime_stats['SFPD MAD'] = person_MAD_sfpd
person_crime_stats['Twitter Slope'] = person_slope_twitter
person_crime_stats['SFPD Slope'] = person_slope_sfpd
```

```python
property_crime_stats = pd.DataFrame({'Month' : months})
property_crime_stats['Twitter MAD'] = property_MAD_twitter
property_crime_stats['SFPD MAD'] = property_MAD_sfpd
property_crime_stats['Twitter Slope'] = property_slope_twitter
property_crime_stats['SFPD Slope'] = property_slope_sfpd
```

```python
drug_crime_stats = pd.DataFrame({'Month' : months})
drug_crime_stats['Twitter MAD'] = drug_MAD_twitter
drug_crime_stats['SFPD MAD'] = drug_MAD_sfpd
drug_crime_stats['Twitter Slope'] = drug_slope_twitter
drug_crime_stats['SFPD Slope'] = drug_slope_sfpd
```

```python
minor_crime_stats = pd.DataFrame({'Month' : months})
minor_crime_stats['Twitter MAD'] = minor_MAD_twitter
minor_crime_stats['SFPD MAD'] = minor_MAD_sfpd
minor_crime_stats['Twitter Slope'] = minor_slope_twitter
minor_crime_stats['SFPD Slope'] = minor_slope_sfpd
```

# plots of normalized mean absolute deviation

```python
plt.plot(person_crime_stats['Month'], person_crime_stats['Twitter MAD'],
        label="Twitter")
plt.plot(person_crime_stats['Month'], person_crime_stats['SFPD MAD'],
        label="SFPD")
plt.legend()
plt.ylim(0.1, 1.3)
plt.title('Change in Variance (Person Crime)')
```

```python
plt.plot(drug_crime_stats['Month'], drug_crime_stats['Twitter MAD'],
        label="Twitter")
plt.plot(drug_crime_stats['Month'], drug_crime_stats['SFPD MAD'],
        label="SFPD")
plt.legend()
plt.ylim(0.1, 1.3)
plt.title('Change in Variance (Drug Crime)')
```

```python
plt.plot(minor_crime_stats['Month'], minor_crime_stats['Twitter MAD'],
        label="Twitter")
plt.plot(minor_crime_stats['Month'], minor_crime_stats['SFPD MAD'],
        label="SFPD")
plt.legend()
plt.ylim(0.1, 1.3)
plt.title('Change in Variance (Minor Offense)')
```

```python
crime_types = ['Person Crime', 'Property Crime', 'Drug Crime', 'Minor Offense']
twitter_slopes = [0.00022701, 0.00027414, -0.00011566, 0.00057015]
sfpd_slopes = [-2.29121053e-06, -0.00369915, -0.00877604, -0.00080263]
twitter_mad = [0.639760, 0.796101, 0.720357, 0.394159]
sfpd_mad = [0.874268, 0.838314, 0.532240, 0.749431]
```

```python
overall_crime_stats = pd.DataFrame({'Crime Types' : crime_types})
overall_crime_stats['Twitter MAD'] = twitter_mad
overall_crime_stats['SFPD MAD'] = sfpd_mad
overall_crime_stats['Twitter Slope'] = twitter_slopes
overall_crime_stats['SFPD Slope'] = sfpd_slopes
```

```python
plt.plot(overall_crime_stats['Crime Types'], overall_crime_stats['Twitter Slope'],
        label="Twitter")
plt.plot(overall_crime_stats['Crime Types'], overall_crime_stats['SFPD Slope'],
        label="SFPD")
plt.legend()
#plt.ylim(-0.05, 0.05)
plt.title('Overall Slopes')
```

```python

plt.plot(overall_crime_stats['Crime Types'], overall_crime_stats['Twitter Slope'],
        label="Twitter")
plt.plot(overall_crime_stats['Crime Types'], overall_crime_stats['SFPD Slope'],
        label="SFPD")
plt.legend()
#plt.ylim(-0.05, 0.05)
plt.title('Overall Slopes')
```

```python
import numpy as np
X_axis = np.arange(len(overall_crime_stats['Crime Types']))
  
plt.bar(X_axis - 0.2, overall_crime_stats['Twitter Slope'], 0.4, label = 'Twitter')
plt.bar(X_axis + 0.2, overall_crime_stats['SFPD Slope'], 0.4, label = 'SFPD')

plt.xticks(X_axis, overall_crime_stats['Crime Types'])
plt.xlabel("Crime Types")
plt.ylabel("Slopes (Normalized)")
plt.title("Slopes over 3 Months (Feb, March, Apr)")
plt.legend()
plt.show()
```
